### Importing packages

```python
import pandas as pd
from matplotlib import pyplot as plt
```

### Reading csv file

```python
healthcare = pd.read_csv("healthcare.csv")
healthcare.head()
```

### Viewing different diagnosis in dataset

```python
healthcare['DRG Definition'].unique()
```

### Selecting rows where diagnosis is Chest Pain

```python
chest_pain = healthcare[healthcare['DRG Definition'] == '313 - CHEST PAIN']
chest_pain.head()
```

### Separating Chest Pain according to State (Alabama)

```python
alabama_chest_pain = chest_pain[chest_pain['Provider State'] == 'AL']
alabama_chest_pain.head()
```

### Obtaining charges of Chest Pain in Alabama

```python
costs = alabama_chest_pain[' Average Covered Charges '].values
costs
```

### Boxplot of Alabama Charges for Chest Pain

```python
plt.boxplot(costs, labels = ['Alabama'])
plt.title('Boxplot of Alabama Charges for Chest Pain')
plt.ylabel('Costs')
plt.show()
```

### Viewing different states in Chest Pain dataset

```python
states = chest_pain['Provider State'].unique()
states
```

### Boxplot of Charges of Chest Pain in different States

```python
dataset = chest_pain[['Provider State', ' Average Covered Charges ']]
datasets = []
for state in states:
    datasets.append(chest_pain[chest_pain['Provider State'] == state][' Average Covered Charges '].values)

plt.figure(figsize=(20,6))
plt.boxplot(datasets, labels = states)
plt.title('Boxplot of Charges of Chest Pain in different States')
plt.xlabel('State')
plt.ylabel('Cost')
plt.show()
```
